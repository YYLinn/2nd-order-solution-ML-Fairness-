# Visualizations for Slides

```python
import matplotlib.pyplot as plt

# Define plotting parameters
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.style.use('ggplot')
```

```python
# Data for the stacked bar chart
labels = ['White', 'Non-White']
approved = [40, 30]  # Number of approved loans
total = [50, 50]  # Total applications
not_approved = [total[i] - approved[i] for i in range(len(approved))]  # Calculating not approved

# Proportions for annotations
proportions = [f"Approval Probability \n {appr/tot * 100}%" for appr, tot in zip(approved, total)]

# Plotting the stacked bar chart
fig, ax = plt.subplots()
ax.bar(labels, approved, label='Approved', color = ['purple', 'orange'])
ax.bar(labels, not_approved, bottom=approved, label='Not Approved', color = ['purple', 'orange'], alpha = 0.3)

# Adding annotations
for i, label in enumerate(labels):
    plt.text(i, approved[i]+1, proportions[i], ha='center')

# Adding legend and labels
plt.ylabel('Number of Applicants')
plt.title('Loan Approval by Group')
handles = [plt.Rectangle((0,0),1,1, color='black'), plt.Rectangle((0,0),1,1, color='black', alpha = 0.3)]
plt.legend(handles, ["Approved", "Not Approved"])
plt.show()
```

```python
# Data for the stacked bar chart
labels = ['White', 'Non-White']
approved = [26, 23]  # Number of approved loans
total = [40, 40]  # Total applications
not_approved = [total[i] - approved[i] for i in range(len(approved))]  # Calculating not approved

# Proportions for annotations
proportions = [f"Approval Probability \n {appr/tot * 100:.1f}%" for appr, tot in zip(approved, total)]

# Plotting the stacked bar chart
fig, ax = plt.subplots()
ax.bar(labels, approved, label='Approved', color = ['purple', 'orange'])
ax.bar(labels, not_approved, bottom=approved, label='Not Approved', color = ['purple', 'orange'], alpha = 0.3)

# Adding annotations
for i, label in enumerate(labels):
    plt.text(i, approved[i]+1, proportions[i], ha='center')

# Adding legend and labels
plt.ylabel('Number of Applicants')
plt.title('Loan Approval by After Controlling for Income and Education')
handles = [plt.Rectangle((0,0),1,1, color='black'), plt.Rectangle((0,0),1,1, color='black', alpha = 0.3)]
plt.legend(handles, ["Approved", "Not Approved"])
plt.show()
```
